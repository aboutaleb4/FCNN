### Layers

```python
class Weight_Layer:
    def __init__(self, input_size, output_size):
        self.input_size = input_size
        self.output_size = output_size
        self.W = np.random.randn(input_size, output_size) / np.sqrt(input_size)
        self.B = np.random.randn(1, output_size) / np.sqrt(input_size)
        self.dW = np.random.randn(input_size, output_size)
        self.dW = self.dW - self.dW
        self.dB = np.random.randn(1, output_size)
        self.dB = self.dB - self.dB
        self.i = 0

    def forward(self, input):
        self.input = input
        return np.dot(input, self.W) + self.B

    def backward(self, batch_size, dOutput, learning_rate):
#         self.i += 1
        dInput = np.dot(dOutput, self.W.T)
        self.dW = np.dot(self.input.T, dOutput)
        self.dB += dOutput 

#         if self.i % batch_size == 0:
        self.dW = self.dW / batch_size
        self.dB = self.dB / batch_size
        self.W -= learning_rate * self.dW
        self.B -= learning_rate * self.dB
        self.dW = self.dW - self.dW
        self.dB = self.dB - self.dB
            
        return dInput
```

```python
class ReLU_Layer:
    
    def forward(self, input):
        self.input = input
        return np.maximum(input, 0)
    
    def backward(self, batch_size, dOutput, learning_rate):
        return dOutput * np.array(self.input >= 0).astype('int')
```

### Network

```python
    
class Network:
    def __init__(self):
        self.network = [
            Weight_Layer(3072, 1000),
            ReLU_Layer(),
            Weight_Layer(1000, 1000),
            ReLU_Layer(),
            Weight_Layer(1000, 5)
        ]

    def test(self, test_set, test_labels):
        num_test = test_set.shape[0]
        Ypred = np.zeros(num_test, dtype = test_labels.dtype)
        for i, x in enumerate(test_set):
            output = np.array([x])
            for layer in self.network:
                output = layer.forward(output)
            Ypred[i] = np.argmax(output)
        accuracy = np.mean(Ypred == test_labels)
        return accuracy
        
    def train(self, epochs, batch_size, train_set, train_labels):
        loss = 0
        loss_history = []
        learning_rate = 0.001
        num_batches = 2000/batch_size
        for epoch in range(epochs):
            loss = 0
            error = 0
            for i in range(num_batches):
                x = train_set[batch_size*i:batch_size*(i+1)]
                y_true = train[batch_size*i:batch_size*(i+1)]
                # forward
                output = x
                for layer in self.network:
                    output = layer.forward(output)

                exp_scores = np.exp(output)
                probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True) # [N x K]
                correct_logprobs = -np.log(probs[0][y_true])
                loss += correct_logprobs
                
                dscores = probs
                dscores[0][y_true] -= 1

                # backward
                output_error = dscores
                for layer in reversed(self.network):
                    output_error = layer.backward(batch_size, output_error, learning_rate)
                    
            train_acc = self.test(train_set[0:2000], train_labels[0:2000])
            val_acc = self.test(train_set[2000:2500], train_labels[2000:2500])
            loss /= 2000
            print(epoch + 1, epochs, train_acc, val_acc)     
            
            #Appending a loss value
            loss_history.append(loss)
                    
        plt.plot(loss_history)
        plt.title('Loss History')
        plt.ylabel('Loss')
        plt.xlabel('epoch')
        plt.show()
```

```python
net = Network()
net.train(200, 100, train_set, train_labels)
```

```python
train_acc = net.test(train_set, train_labels)
print(train_acc)
test_acc = net.test(test_set, test_labels)
print(test_acc)
```
