```python
import numpy as np

def L_i_vectorized(scores, y):
 delta = 1.0
 # compute the margins for all classes in one vector operation
 margins = np.maximum(0, scores - scores[y] + delta)
 # on y-th position scores[y] - scores[y] canceled and gave delta. We want
 # to ignore the y-th position and only consider margin on max wrong class
 margins[y] = 0
 loss_i = np.sum(margins)
 return loss_i

def LeakyReLU(x):
    return max(x, 0.01*x)

def ReLU(x):
    a = x
    a[a < 0] = 0
    return a

class FCNN(object):
    
  def __init__(self, input_size = 3073, output_size = 5, NumberOfLayers = 2, NeuronsPerLayer = 50):
    self.NumberOfLayers = NumberOfLayers
    self.NeuronsPerLayer = NeuronsPerLayer
    self.input_size = input_size
    self.output_size = output_size
    self.W = []
    self.W.append(np.random.randn(input_size,NeuronsPerLayer)/np.sqrt(input_size))
    for i in range(NumberOfLayers - 1):
        self.W.append(np.random.randn(NeuronsPerLayer,NeuronsPerLayer)/np.sqrt(NeuronsPerLayer))
    self.W.append(np.random.randn(NeuronsPerLayer,output_size)/np.sqrt(NeuronsPerLayer))
#     for i in self.W:
#         print(i.shape)
#     self.printInfo()

  def printInfo(self):
    print (self.input_size)
    print (self.output_size)
    print (self.NumberOfLayers)
    print (self.NeuronsPerLayer)
    
  def forward(self, X, Y):
    h = []
    h.append(ReLU(np.dot(X, self.W[0])))
    for i in range(self.NumberOfLayers - 1):
        h.append(ReLU(np.dot(h[i], self.W[i + 1])))
    scores = ReLU(np.dot(h[self.NumberOfLayers - 1], self.W[self.NumberOfLayers]))
    loss = L_i_vectorized(scores, Y)
    return scores

  def backward(self):
    dScores = np.zeros(self.output_size)
    dW = []
    dW.append(np.zeros(self.input_size, self.NeuronsPerLayer))
    for i in range(self.NumberOfLayers - 1):
        dW.append(np.zeros(self.NeuronsPerLayer,self.NeuronsPerLayer))
    dW.append(np.zeros(self.NeuronsPerLayer,self.output_size))
    dW = np.array(dW)
    print(dW.shape)
#     return dScores, dW

  def train(self, X, Y):
    num_test = X.shape[0]
    Ypred = np.zeros(num_test, dtype = Y.dtype)
    for i, row in enumerate(X): 
        scores_test = self.forward(row, Y[i])
        Ypred[i] = np.argmax(scores_test)
    self.backward()
    
  def test(self, X, Y):
    num_test = X.shape[0]
    Ypred = np.zeros(num_test, dtype = Y.dtype)
    for i, row in enumerate(X): 
        scores_test = self.forward(row, Y[i])
        Ypred[i] = np.argmax(scores_test)
    num_correct = np.sum(Ypred == Y)
    accuracy = float(num_correct) / 500
    print(accuracy)
```
